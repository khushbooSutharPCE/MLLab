```python
X = np.array(X)
y = np.array(y)
```

```python
def candidate_elimination(X, y):
    num_attributes = len(X[0])
    S = [""] * num_attributes  # Initialize S as a list of ""
    G = [["?"] * num_attributes]  # Initialize G as a list containing a list of "?"

    for i, instance in enumerate(X):
        if y[i] == "yes":
            for j in range(num_attributes):
                if S[j] == "":
                    S[j] = instance[j]
                elif S[j] != instance[j]:
                    S[j] = "?"

            # Ensure G elements have the same length as S (This part is likely unnecessary now)
            # G = [g if len(g) == num_attributes else g + ["?"] * (num_attributes - len(g)) for g in G]

            G = [g for g in G if all(
                g[k] == "?" or S[k] == "?" or g[k] == S[k]
                for k in range(min(len(g), len(S)))
            )]

        elif y[i] == "no":
            new_G = []
            for g in G:
                for j in range(num_attributes):
                    if len(g) > j and g[j] == '?':  # Check if j is within bounds
                        if S[j] != '?':
                            new_hypothesis = g[:] # Create a copy of g
                            new_hypothesis[j] = S[j]
                            new_G.append(new_hypothesis)

            G = new_G

            G = [g for g in G if any(
                g[k] != instance[k] and g[k] != '?'
                for k in range(min(len(g), len(instance)))
            )]

    return S, G
```

```python
S_final, G_final = candidate_elimination(X, y)
print("Final Specific Hypothesis:", S_final)
print("Final General Hypothesis:", G_final)
```
